```python
from datasets import load_dataset
qna_dataset = load_dataset("sadeem-ai/arabic-qna")
news_dataset = load_dataset("arbml/SANAD")
```

```python
print(qna_dataset)
```

```python
qna_dataset = qna_dataset.filter(lambda exemple: exemple["has_answer"] == True)
```

```python
qna_dataset
```

```python
doc_texts = qna_dataset["train"]["text"]
```

```python
for i in range(len(doc_texts)):
  print(doc_texts[i])
```

```python
metadata = [
    {
        "source" : rec["source"],
        "title" : rec["title"]
    }
    for rec in qna_dataset["train"]
]
metadata
```

```python
docs_ids = [
    str(i)
    for i in range(len(doc_texts))
]
docs_ids
```

```python
from sentence_transformers import SentenceTransformer

model_id = "sentence-transformers/distiluse-base-multilingual-cased-v2"
device = "cuda:0" # "cpu"
dim = 512

model = SentenceTransformer(model_id , device=device)
```

```python
text  = "السَّلَامُ عَلَيْكُمْ وَرَحْمَةُ ٱللَّٰهِ وَبَرَكَاتُهُ"

text_embed = model.encode(text)
print(text_embed.shape)
print(text_embed)
```

```python
encoded_docs = model.encode(list(doc_texts), show_progress_bar=True)
```

```python
encoded_docs.shape
```

```python
import chromadb

chroma_client = chromadb.PersistentClient(path="./chromadb-ar-docs")
```

```python
collection  = chroma_client.create_collection(
    name="ar_docs",
    metadata={"hnsw:space": "cosine"}
)
```

```python
collection.add(
    documents=doc_texts,
    embeddings=encoded_docs.tolist(),
    metadatas=metadata,
    ids=docs_ids,
)
```

```python
question = "ما السبب في صغر الأسنان بالمقارنة مع حجم الفكين؟"

question_embed = model.encode(question)
result = collection.query(
    query_embeddings=question_embed.tolist(),
    n_results=3
)

print(result)
```
